### Самый популярный объект

Самый популярный пост -- это тот, с которым зафиксировано максимальное число взаимодействий (комментарии, репосты и лайки, т.е. информация, представленная в колонке Total interactions, а также максимальный охват (Lifetime post total reach, кол-во уникальных пользователей, увидевших пост) и число показов (Lifetime post total impressions, кол-во неуникальных пользователей, просмотревших пост). Сначала посмотрим на кол-во показов и охват:

```python
df[(df['Lifetime Post Total Reach'] == df['Lifetime Post Total Reach'].max())] 
```

```python
df[(df['Lifetime Post Total Impressions'] == df['Lifetime Post Total Impressions'].max())] 
```

Видно, что это разные посты. У первого больше охват, а у второго -- число показов (его смотрели много раз одни и те же пользователи и, как видно по столбцу Lifetime Post Impressions by people who have liked your Page, это те, кто лайкал ранее посты на данной странице, то есть, скорее всего, аудитория, которой нравится весь контент страницы в целом). А в первом посте больше показов от людей, которые еще не лайкали постов на странице.
Теперь нужно проверить, какой пост из датасета активнее всего лайкали, репостили и комментировали. 

```python
df[(df['Total Interactions'] == df['Total Interactions'].max())]
```

Видно, что это как раз пост с наибольшим охватом. Цифры в столбцах взаимодействий с постом  в разы больше по сравнению с постом с наибольшим числом показов. Он является самым популярным объектов в выборке.
